```python
import json
import random
import PIL
import datetime

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python

  
# Opening JSON file
f = open('tarot-images.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
cards = data['cards']
```

```python
#  https://stackoverflow.com/questions/22842289/generate-n-unique-random-numbers-within-a-range

random_cards = random.sample(range(0, 77), 3)
random_cards
```

```python
past_dict = cards[random_cards[0]]
present_dict = cards[random_cards[1]]
future_dict = cards[random_cards[2]]
```

```python
print(f"Past: {past_dict['name']}")
print(f"Present: {present_dict['name']}")
print(f"Future: {future_dict['name']}")
```

```python
    # identify images
name_img_past = past_dict['img']
name_img_present = present_dict['img']
name_img_future = future_dict['img']
```

```python
#https://www.kaggle.com/code/valeriaf22/tarotdecknotebook

img_past = PIL.Image.open(f'cards/{name_img_past}')
img_present = PIL.Image.open(f'cards/{name_img_present}')
img_future = PIL.Image.open(f'cards/{name_img_future}')

# plot images
fig, (past, present, future) = plt.subplots(1, 3, figsize=(10,5.5))

#fig.suptitle('Your reading: Past, Present, Future on {}'.format(date))
fig.suptitle('Your reading: Past, Present, Future on ')

past.imshow(img_past)
past.axis('off')
past.set_title(past_dict['name'])
present.imshow(img_present)
present.axis('off')
present.set_title(present_dict['name'])
future.imshow(img_future)
future.axis('off')
future.set_title(future_dict['name'])
plt.show()
```
